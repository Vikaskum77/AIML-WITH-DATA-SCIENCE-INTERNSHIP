```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 # IPL dataset (2021-2024)
data = {
    'season': [2024, 2024, 2024, 2023, 2023, 2023, 2022, 2022, 2022, 2021, 2021, 2021],
    'team1': ['KKR', 'DC', 'MI', 'MI', 'CSK', 'RCB', 'MI', 'KKR', 'CSK', 'DC', 'RCB', 'MI'],
    'team2': ['RCB', 'CSK', 'DC', 'CSK', 'RCB', 'MI', 'KKR', 'CSK', 'MI', 'MI', 'CSK', 'KKR'],
    'winner': ['KKR', 'CSK', 'MI', 'MI', 'CSK', 'MI', 'MI', 'KKR', 'CSK', 'DC', 'RCB', 'MI'],
    'city': ['Kolkata', 'Delhi', 'Mumbai', 'Mumbai', 'Chennai', 'Bangalore',
             'Kolkata', 'Chennai', 'Mumbai', 'Delhi', 'Bangalore', 'Kolkata'],
    'player_of_match': ['Shreyas Iyer', 'David Warner', 'Ishan Kishan',
                        'Rohit Sharma', 'MS Dhoni', 'Virat Kohli', 'Jasprit Bumrah',
                        'Andre Russell', 'Ravindra Jadeja', 'Rishabh Pant',
                        'AB de Villiers', 'Kieron Pollard'],
    'runs_scored': [200, 165, 180, 160, 180, 175, 150, 200, 190, 170, 185, 165]
}

ipl_df = pd.DataFrame(data)

# Display the dataset
print(ipl_df)
```

```python
# 1. Basic Data Exploration

print(ipl_df.head())  # View the first few rows
print(ipl_df.info())  # Get information about columns and data types
```

```python
# 2. Team Performance
team_wins = ipl_df['winner'].value_counts()
print("\nTeam Wins:\n", team_wins)
```

```python
# 3. Player of the Match Analysis
player_of_match_counts = ipl_df['player_of_match'].value_counts()
print("\nPlayer of the Match Counts:\n", player_of_match_counts)
```

```python
# 4. Runs Scored Analysis
average_runs_per_match = ipl_df['runs_scored'].mean()
print("\nAverage Runs Scored per Match:", average_runs_per_match)
```

```python
# 1. Bar Chart of Team Wins

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")

plt.figure(figsize=(10, 6))
sns.countplot(x='winner', data=ipl_df, hue='winner', palette='viridis', legend=False)
plt.title('Number of Wins by Team', fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel('Team', fontsize=12, fontweight='bold', color='darkred')
plt.ylabel('Number of Wins', fontsize=12, fontweight='bold', color='darkgreen')

plt.show()
```

```python
# 2. Histogram of Runs Scored
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")

plt.figure(figsize=(10, 6))
plt.hist(ipl_df['runs_scored'], bins=8, color='royalblue', edgecolor='black', alpha=0.7)  # Enhanced colors
plt.title('Distribution of Runs Scored per Match', fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel('Runs Scored', fontsize=12, fontweight='bold', color='darkred')
plt.ylabel('Frequency', fontsize=12, fontweight='bold', color='darkgreen')

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
