```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Air Quality Data for Indian Cities (Extended for 2024 and 2025)
data = {
    'city': ['Delhi', 'Mumbai', 'Kolkata', 'Chennai', 'Bangalore', 'Hyderabad', 'Pune', 'Ahmedabad', 'Jaipur', 'Lucknow'] * 3,
    'date': pd.to_datetime([
        '2023-01-15', '2023-01-15', '2023-01-15', '2023-01-15', '2023-01-15',
        '2023-01-15', '2023-01-15', '2023-01-15', '2023-01-15', '2023-01-15',
        '2024-01-15', '2024-01-15', '2024-01-15', '2024-01-15', '2024-01-15',
        '2024-01-15', '2024-01-15', '2024-01-15', '2024-01-15', '2024-01-15',
        '2025-01-15', '2025-01-15', '2025-01-15', '2025-01-15', '2025-01-15',
        '2025-01-15', '2025-01-15', '2025-01-15', '2025-01-15', '2025-01-15'
    ]),
    'pm25': [250, 120, 180, 90, 70, 150, 110, 160, 200, 140,
             260, 130, 190, 100, 80, 160, 120, 170, 210, 150,
             270, 140, 200, 110, 90, 170, 130, 180, 220, 160],
    'no2': [80, 50, 60, 40, 30, 70, 45, 65, 90, 55,
            85, 55, 65, 45, 35, 75, 50, 70, 95, 60,
            90, 60, 70, 50, 40, 80, 55, 75, 100, 65],
    'so2': [36, 20, 25, 15, 10, 25, 18, 23, 35, 22,
            38, 22, 27, 17, 12, 27, 20, 25, 37, 24,
            40, 24, 29, 19, 14, 29, 22, 27, 39, 26]
}

air_quality_df = pd.DataFrame(data)

# Display the first few rows
print(air_quality_df.head(15))
```

```python
# 1. Basic Data Exploration
print(air_quality_df.head())  # Display the first few rows of the dataset
print(air_quality_df.info())  # Display dataset information (column names, data types, non-null values)
```

```python
# 2. PM2.5 Analysis
average_pm25 = air_quality_df['pm25'].mean()  # Calculate the mean of PM2.5 levels
print("\nAverage PM2.5 across cities:", average_pm25)  # Print the average value
```

```python
# 3. City with Highest NO2 Levels
city_highest_no2 = air_quality_df.loc[air_quality_df['no2'].idxmax(), 'city']
print("\nCity with highest NO2 levels:", city_highest_no2)
```

```python
# 4. Correlation between Pollutants
correlation_pm25_no2 = air_quality_df['pm25'].corr(air_quality_df['no2'])
print("\nCorrelation between PM2.5 and NO2:", correlation_pm25_no2)
```

```python
# 1. Bar Chart of PM2.5 Levels by City
plt.figure(figsize=(12, 6))  # Set figure size
sns.barplot(x='city', y='pm25', data=air_quality_df)  # Create bar plot
plt.title('PM2.5 Levels by City')  # Add title
plt.xlabel('City')  # Label x-axis
plt.ylabel('PM2.5')  # Label y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()  # Display plot
```

```python
# 2. Scatter Plot of PM2.5 vs. NO2
plt.figure(figsize=(8, 6))  # Set figure size
sns.scatterplot(x='pm25', y='no2', data=air_quality_df, hue='city')  # Scatter plot with city-based color coding
plt.title('PM2.5 vs. NO2 Levels')  # Add title
plt.xlabel('PM2.5')  # Label x-axis
plt.ylabel('NO2')  # Label y-axis
plt.show()  # Display plot
```
